Scaling of Features is an essential step in modeling the algorithms with the datasets.

The data obtained contains features of various dimensions and scales altogether. Different scales of the data 
features affect the modeling of a dataset adversely.

It leads to a biased outcome of predictions in terms of misclassification error and accuracy rates. Thus, it is
necessary  to scale the data prior to modeling.

This is when standardization comes into picture.

Standardization is a scaling technique wherein it makes the data scale-free by converting the statistical 
distribution of the data into the below format:
    
mean - 0 (zero)
standard deviation - 1

Z= (X-Mean)/SD

# Principal Component Analysis (2 dimensions)



```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)

i = pca.fit_transform(X_train,y_train)
i
principal_df = pd.DataFrame(data=i,columns= ['PC1','PC2'])
principal_df
```

# Clustering - FInd the optimal number of clusters - Elbow method 

Plot Sum of Squared Errors (SSE)against K value 

```python
from sklearn.cluster import KMeans

SSE = []

K= range(2,9)

for i in K:
    kmeanModel = KMeans(n_clusters=i)
    kmeanModel.fit(principal_df)
    SSE.append(kmeanModel.inertia_)
```

```python
plt.figure()
plt.plot(K,SSE)

plt.ylabel('Inertia(SSE)')
plt.xlabel('Number of clusters(K)')
plt.show()
```

We use K=3 ; K Means 

```python
kmeans = KMeans(n_clusters = 3).fit(principal_df)
plt.scatter(principal_df['PC1'],principal_df['PC2'],c=kmeans.labels_)
plt.title('Diabetes 2D K-Means')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(linestyle='dotted')
plt.show()
```
